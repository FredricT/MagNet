```python
class_freqs = {
    "sin\n1k": I_sin_1k.shape[0],
    # "V_sin_1k": V_sin_1k.shape[0],
    "tri\n1k": I_tri_1k.shape[0],
    # "V_tri_1k": V_tri_1k.shape[0],
    "trap\n1k": I_trap_1k.shape[0],
    # "V_trap_1k": V_trap_1k.shape[0],
    "sin\n2k": I_sin_2k.shape[0],
    # "V_sin_2k": V_sin_2k.shape[0],
    "tri\n2k": I_tri_2k.shape[0],
    # "V_tri_2k": V_tri_2k.shape[0],
    "trap\n2k": I_trap_2k.shape[0],
    # "V_trap_2k": V_trap_2k.shape[0],
    "sin\n5k": I_sin_5k.shape[0],
    # "V_sin_5k": V_sin_5k.shape[0],
    "tri\n5k": I_tri_5k.shape[0],
    # "V_tri_5k": V_tri_5k.shape[0],
    "trap\n5k": I_trap_5k.shape[0],
    # "V_trap_5k": V_trap_5k.shape[0],
}
class_freqs = pd.DataFrame.from_dict(class_freqs, orient="index", columns=["frequency"])
class_freqs = class_freqs.sort_values(by=["frequency"], ascending=False)
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(12, 8)
ax.bar(x=class_freqs.index, height=class_freqs.frequency)
ax.tick_params(labelright=False, labelsize=24)
# for tick in ax.get_xticklabels(): tick.set_rotation(90)
ax.set_title("Data quantity for each wave type", fontsize=32)
# ax.set_xlabel("Wave Shape and Fundamental Frequency", fontsize=32)
# ax.set_ylabel("Quantity", fontsize=32)
ax.set_yticks(np.append(np.linspace(0, 2000, 2), np.mean(class_freqs.frequency)))
ax.set_yticklabels([ str(int(i)) for i in ax.get_yticks()])

ax.axhline(np.mean(class_freqs.frequency), color="black", linestyle="--")
fig.savefig("classes.eps")
fig.show()
```
